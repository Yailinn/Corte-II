#Feigenbaum and logistic map

Calcular las primeras mil iteraciones

```python
from pylab import*
from numpy import*
x_0=1/2
r=1.5
for i in range(1000):
    x_0=r*x_0*(1-x_0)
    #print(x_0)
    plot(i,x_0,"ro")
    grid("on")
```

```python
x_0=1/2
r=2
for i in range(1000):
    x_0=r*x_0*(1-x_0)
    #print(x_0)
    plot(i,x_0,"go")
    grid("on")
```

```python
from numpy import*
x_0=1/2
for r in [float(j)/100 for j in range(100,401,1)]:
    
    for i in range(50):
        x_0=r*x_0*(1-x_0)
        #print(x_0)
        plot(r,x_0,"ro")
        grid("on")
        #print(r)
show()
```
